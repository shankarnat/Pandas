```python
#Regex and Pandas

import pandas as pd

time_sentences = ["Monday: The doctor's appointment is at 2:45pm.", 
                  "Tuesday: The dentist's appointment is at 11:30 am.",
                  "Wednesday: At 7:00pm, there is a basketball game!",
                  "Thursday: Be back home by 11:15 pm at the latest.",
                  "Friday: Take the train at 08:10 am, arrive at 09:00am."]

df = pd.DataFrame(time_sentences, columns=['text'])
df
```

```python
#Get the rows has appointment only
df[df['text'].str.contains('appointment')]
```

```python
#find number of rows that has appointment - could be interesting number to get as well
len(df[df['text'].str.contains('appointment')])
```

```python
#extract the time
df['time'] = df['text'].str.extract(r'(\d?\d:\d\d\s?[a-z]{2})')
```

```python
df
```

```python
df['numlen'] = df['text'].str.count(r'\d')
```
